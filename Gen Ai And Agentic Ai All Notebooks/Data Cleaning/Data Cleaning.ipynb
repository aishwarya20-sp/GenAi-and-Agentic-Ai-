# Data Cleaning

#  1)strip()

### strip() is a string method used in data cleaning to remove extra spaces or unwanted characters from the beginning and end of a text value.

```python
text="\t\t\t\t\nhello   world"
text.strip()
```

```python
name = "   Aishwarya   "
clean_name=name.strip()
clean_name
```

## 1. lstrip()

## lstrip() removes spaces from the left side / beginning of text.

```python
text="\t\t\t\t\nhello   world\t\t\t"
text.lstrip()
```

```python
text = "   Aishwarya"

clean_text = text.lstrip()

print(clean_text)
```

## 2. rstrip()

## rstrip() removes spaces from the right side / end of text.

```python
text="\t\t\t\t\nhello   world\t\t\t"
text.rstrip()
```

```python
text = "Aishwarya    "
clean_text=text.rstrip()
clean_text
```

# 2)replace

## replace() is a string method used to replace one text/character with another text/character.

```python
text="hello   world hello"
clean_text=text.replace("hello","hii")
clean_text
```

```python
text = "I like Java"

new_text = text.replace("Java", "Python")

print(new_text)
```

```python
text="\t\t\t\t\nhello   world\t\t\t"
clean_text=text.replace("\t"," ")
clean_text# we can replace as a space
```

```python
text="\t\t\t\t\nhello   world\t\t\t"
clean_text=text.replace("\t","")
clean_text
```

## Why regex is used in data cleaning?

### Regex is useful for cleaning text like:

### remove special characters
### extract phone numbers
### extract email IDs
### remove extra spaces
### replace unwanted symbols
### validate formats

```python
 1) \d= digit
 2) \D= not digit
 3) \w= char ( alphanum _)
 4) \W= not char ( alphanum _)
 5) \s= space tab newline
 6) \S= not space tab newline
 7) .= any char
 8) + = one or more
 9) * = zero or more
 10)? = optional
 11) {n}= exact count
 12) [] = set char
 13) [a-z]= small letters
 14) [A-Z]= upper case
 15) [0-9]= numbers
 16) [a-zA-Z]= all aplhabets
 17) [a-zA-Z0-9]= alphanumeric
 18) ()= group
 19) | = or
 20) ^ = startwith
 21) $ = end
  re.search()=  first match
  re.findall()= all matches
  re.match()= match from start
  re.fullmatch()= entire string mathc
  re.sub()= replace
```
