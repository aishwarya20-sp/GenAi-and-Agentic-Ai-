## Read text/content from every file.

```python
import os
import json
import csv
import fitz   # PyMuPDF
# insert files
files = os.listdir()

for i in files:
    print(i)

    if i.endswith(".txt"):
        print("TXT")

        def txt_read(files):
            text = open(files, "r", encoding="utf-8")
            data = text.read()
            text.close()
            return data

        data = txt_read(i)
        print(data)

    elif i.endswith(".json"):
        print("JSON")

        def json_read(files):
            text = open(files, "r", encoding="utf-8")
            data = json.load(text)
            text.close()
            return data

        data = json_read(i)
        print(data)

    elif i.endswith(".csv"):
        print("CSV")

        def csv_read(files):
            text = open(files, "r", encoding="utf-8")
            reader = csv.DictReader(text)

            data = ""
            for row in reader:
                data += str(row) + "\n"

            text.close()
            return data

        data = csv_read(i)
        print(data)

    elif i.endswith(".pdf"):
        print("PDF")

        def pdf_read(files):
            doc = fitz.open(files)

            all_text = ""
            for page in doc:
               all_text += page.get_text()

            doc.close()
            return data

        data = pdf_read(i)
        print(data)

    else:
        print("FILES IS UNSUPPORTED")
```

## Append all extracted text into a single variable named all_text.

## Handle unsupported file types by skipping them.

```python
import os
import json
import csv
import fitz   # PyMuPDF

files = os.listdir()

all_text = ""

for i in files:
    print(i)

    if i.lower().endswith(".txt"):
        print("TXT")

        def txt_read(files):
            text = open(files, "r", encoding="utf-8")
            data = text.read()
            text.close()
            return data

        data = txt_read(i)
        all_text += data + "\n"

    elif i.lower().endswith(".json"):
        print("JSON")

        def json_read(files):
            text = open(files, "r", encoding="utf-8")
            data = json.load(text)
            text.close()
            return str(data)

        data = json_read(i)
        all_text += data + "\n"

    elif i.lower().endswith(".csv"):
        print("CSV")

        def csv_read(files):
            text = open(files, "r", encoding="utf-8")
            reader = csv.DictReader(text)

            data = ""
            for row in reader:
                data += str(row) + "\n"

            text.close()
            return data

        data = csv_read(i)
        all_text += data + "\n"

    elif i.lower().endswith(".pdf"):
        print("PDF")

        def pdf_read(files):
            doc = fitz.open(files)

            data = ""
            for page in doc:
                data += page.get_text()

            doc.close()
            return data

        data = pdf_read(i)
        all_text += data + "\n"

    else:
        print("Skipping unsupported file")
        continue

print("\n========== FINAL COMBINED TEXT ==========\n")
print(all_text)
```

## Keep the code modular and reusable.

```python
import os
import json
import csv
import fitz   # PyMuPDF


def txt_read(file_name):
    text = open(file_name, "r", encoding="utf-8")
    data = text.read()
    text.close()
    return data


def json_read(file_name):
    text = open(file_name, "r", encoding="utf-8")
    data = json.load(text)
    text.close()
    return str(data)


def csv_read(file_name):
    text = open(file_name, "r", encoding="utf-8")
    reader = csv.DictReader(text)

    data = ""
    for row in reader:
        data += str(row) + "\n"

    text.close()
    return data


def pdf_read(file_name):
    doc = fitz.open(file_name)

    data = ""
    for page in doc:
        data += page.get_text()

    doc.close()
    return data


def read_file(file_name):
    if file_name.lower().endswith(".txt"):
        print("TXT")
        return txt_read(file_name)

    elif file_name.lower().endswith(".json"):
        print("JSON")
        return json_read(file_name)

    elif file_name.lower().endswith(".csv"):
        print("CSV")
        return csv_read(file_name)

    elif file_name.lower().endswith(".pdf"):
        print("PDF")
        return pdf_read(file_name)

    else:
        print("Skipping unsupported file")
        return None


def read_all_files():
    files = os.listdir()
    all_text = ""

    for i in files:
        print(i)

        # skip folders
        if os.path.isdir(i):
            continue

        data = read_file(i)

        if data is None:
            continue

        all_text += data + "\n"

    return all_text


final_text = read_all_files()

print("\n========== FINAL COMBINED TEXT ==========\n")
print(final_text)
```
