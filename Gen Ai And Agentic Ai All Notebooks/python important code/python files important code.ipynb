## with open("File_name.ext","r=read, w=write,rb=read_binary",encoding="utf-8")

##### Using the with open() syntax in Python is the gold standard for file handling because it acts as a context manager that automatically closes the file once the code block finishes executing

```python
with open("text_file.txt","r",encoding="utf-8") as f:
    test=f.read()#first open the file and read the file and assign as f (whole code) and again stored in new variable as test,encoding=utf-8 for default data
```

```python
print(test)
```

### Direct Open and  Read file

```python
test= open("text_file.txt","r",encoding="utf-8").read()
print(test)
```
